```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/data.lfpn/ibraun/Code/paper_volume_calculation/outputs_patient_data/final_results/SCD0003701/seg_analysis_data/segmentation/seg_volumes_cal.csv")

print(df.head())

mean_vol_bp = df["bp_volume"]
mean_vol_myo = df["myo_volumes"]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example time vector
time_norm = np.linspace(0, 1, len(mean_vol_bp))

# Collect all means for global y-limits
all_means = np.concatenate([
    mean_vol_bp, mean_vol_myo
])
ymin = np.min(all_means) -2
ymax = np.max(all_means) + 2

def plot_volume(ax, time_norm, mean_vol, color_plot, pathology_title):
    """Helper function to plot mean + scatter (no std)."""
    # Plot mean line
    ax.plot(time_norm * 100, mean_vol,
            color=color_plot, linewidth=2.5,
            label="Mean Volume", alpha=0.6, zorder=3)

    # Scatter points
    ax.scatter(time_norm * 100, mean_vol,
               color=color_plot, s=50, alpha=0.5,
               linewidth=1, zorder=4)

    # Labels & title
    ax.set_xlabel('Percentage of Cardiac Cycle [%]', fontsize=16, fontweight='bold', color='#000000')
    ax.set_ylabel('Volume [ml]', fontsize=16, fontweight='bold', color='#000000')
    ax.set_title(f'{pathology_title}', fontsize=18, fontweight='bold',
                 color='#000000', pad=15)

    # Tick styling
    ax.tick_params(axis='both', which='major', labelsize=11, colors='#000000')

    # Spines and grid
    ax.grid(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#000000')
    ax.spines['bottom'].set_color('#000000')

    # Background
    ax.set_facecolor('#FFFFFF')

    # Fix y-limits to be consistent across plots
    ax.set_ylim(ymin, ymax)


# ---- Plot 1: Healthy ----
fig, ax = plt.subplots(figsize=(10,6), dpi=100)
plot_volume(ax, time_norm, mean_vol_bp,
            color_plot="#118ab2ff", pathology_title="Blood Pool Volume")
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.95, top=0.9)
plt.savefig("/data.lfpn/ibraun/Code/paper_volume_calculation/Graphics/example_bp_simpson.pdf",dpi=300)
plt.savefig("/data.lfpn/ibraun/Code/paper_volume_calculation/Graphics/example_bp_simpson.svg")

plt.show()

# ---- Plot 2: Failure Infarct ----
fig, ax = plt.subplots(figsize=(10,6), dpi=100)
plot_volume(ax, time_norm, mean_vol_myo,
            color_plot="#ef476fff", pathology_title="Myocardium Volume")
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.95, top=0.9)
plt.savefig("/data.lfpn/ibraun/Code/paper_volume_calculation/Graphics/example_myo_simpson.pdf",dpi=300)
plt.savefig("/data.lfpn/ibraun/Code/paper_volume_calculation/Graphics/example_myo_simpson.svg")
plt.show()


```
